```python
final_df = pd.merge(new_df, new_df2, on=['Date', 'Sent by'])
```

```python
X = final_df.drop(['Date', 'Network_x_x', 'Post Type', 'Content Type_x', 'Profile', 'Sent by', 'Network_y_x', 'Network_x_y', 'Post Type_x', 'Content Type_y', 'Network_y_y', 'Post Type_y', 'Profile_y'], axis=1)
y = final_df['Engagements']
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
```

```python
lr.fit(X_train, y_train)
```

```python
c = lr.intercept_
c
```

```python
m = lr.coef_
m
```

```python
y_predict_train = lr.predict(X_train)
y_predict_train
```

```python
sns.scatterplot(y_train, y_predict_train)
plt.show()
```

```python
from sklearn.metrics import r2_score
r2_score(y_train, y_predict_train)
```

```python
y_predict_test = lr.predict(X_test)
y_predict_test
```

```python
sns.scatterplot(y_test, y_predict_test)
plt.show()
```

```python
r2_score(y_test, y_predict_test)
```
